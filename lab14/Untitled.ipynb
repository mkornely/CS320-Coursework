```python
from matplotlib import pyplot as plt
img = plt.imread("capital.jpg")
plt.imshow(img)
```

```python
img.shape
```

```python
import pandas as pd
rgb = pd.DataFrame(img.reshape(-1, 3), columns=["red", "green", "blue"])
rgb
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=4)
km.fit(rgb)
```

```python
km.cluster_centers_
```

```python
centroids=pd.DataFrame(km.cluster_centers_.astype(int), columns=rgb.columns)
plt.imshow(centroids.values.reshape(1, 4, 3))
```

```python
km = KMeans(n_clusters=4)
km.fit(img.reshape(-1,3))
plt.imshow(km.cluster_centers_.reshape(1,4,3).astype(int))
```
